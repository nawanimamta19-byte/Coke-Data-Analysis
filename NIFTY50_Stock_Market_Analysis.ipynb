```python

# -----------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

LOAD DATASET

```python
df = pd.read_csv('/content/nifty50_closing_prices.csv')
df
```

```python
print(df.head())
```

DATA UNDERSTANDING

```python
print("\nDataset Information:")
print(df.info())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

DATA CLEANING

```python
# Convert Date column into datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Set Date as index
df.set_index('Date', inplace=True)
```

```python

# Sort values by date
df.sort_index(inplace=True)
```

```python
print("\nCleaned Data:")
print(df.head())
```

📈 TREND ANALYSIS

```python


plt.figure(figsize=(15,6))

for col in df.columns[:5]:
    plt.plot(df.index, df[col], label=col)

plt.title("Stock Price Trend Analysis")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.legend()
plt.show()
```

📉 DAILY RETURNS ANALYSIS

```python

daily_returns = df.pct_change() * 100

print("\nDaily Returns:")
print(daily_returns.head())

# Plot daily returns
plt.figure(figsize=(15,6))

for col in daily_returns.columns[:5]:
    plt.plot(daily_returns.index,
             daily_returns[col],
             label=col)

plt.title("Daily Returns")
plt.xlabel("Date")
plt.ylabel("Return (%)")
plt.legend()
plt.show()
```

📊 MOVING AVERAGE ANALYSIS

```python
stock = df.columns[0]

df['7_Day_MA'] = df[stock].rolling(window=7).mean()
df['30_Day_MA'] = df[stock].rolling(window=30).mean()

plt.figure(figsize=(15,6))

plt.plot(df[stock], label='Original Price')
plt.plot(df['7_Day_MA'], label='7 Day MA')
plt.plot(df['30_Day_MA'], label='30 Day MA')

plt.title(f"Moving Average Analysis - {stock}")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

🔥 CORRELATION ANALYSIS

```python
correlation_matrix = daily_returns.corr()

plt.figure(figsize=(14,10))

sns.heatmap(correlation_matrix,
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Heatmap")
plt.show()
```

📌  VOLATILITY ANALYSIS

```python
volatility = daily_returns.std()

print("\nStock Volatility:")
print(volatility.sort_values(ascending=False))

# Plot volatility
volatility.sort_values().plot(kind='barh',
                              figsize=(10,8))

plt.title("Volatility of Stocks")
plt.xlabel("Standard Deviation")
plt.show()
```

📅  MONTHLY PERFORMANCE ANALYSIS

```python
monthly_prices = df.resample('M').mean()

print("\nMonthly Average Prices:")
print(monthly_prices.head())

# Plot monthly prices
monthly_prices.iloc[:, :5].plot(figsize=(15,6))

plt.title("Monthly Average Stock Prices")
plt.xlabel("Month")
plt.ylabel("Price")
plt.show()
```

📈  CUMULATIVE RETURNS

```python
cumulative_returns = (1 + daily_returns/100).cumprod()

plt.figure(figsize=(15,6))

for col in cumulative_returns.columns[:5]:
    plt.plot(cumulative_returns.index,
             cumulative_returns[col],
             label=col)

plt.title("Cumulative Returns")
plt.xlabel("Date")
plt.ylabel("Growth")
plt.legend()
plt.show()
```

📊 HIGHEST & LOWEST PRICE ANALYSIS

```python
highest_prices = df.max()
lowest_prices = df.min()

print("\nHighest Prices:")
print(highest_prices)

print("\nLowest Prices:")
print(lowest_prices)
```

📉 DISTRIBUTION ANALYSIS

```python

stock_return = daily_returns[df.columns[0]]

plt.figure(figsize=(10,6))

sns.histplot(stock_return.dropna(),
             bins=30,
             kde=True)

plt.title(f"Distribution of Returns - {df.columns[0]}")
plt.xlabel("Daily Return")
plt.show()
```

🚨  OUTLIER DETECTION USING Z-SCORE

```python
from scipy.stats import zscore

z_scores = np.abs(zscore(daily_returns.dropna()))

outliers = (z_scores > 3)

print("\nOutliers Detected:")
print(outliers.sum())
```

📌  RISK VS RETURN ANALYSIS

```python
avg_returns = daily_returns.mean()
risk = daily_returns.std()

risk_return_df = pd.DataFrame({
    'Returns': avg_returns,
    'Risk': risk
})

print("\nRisk vs Return:")
print(risk_return_df)

# Scatter Plot
plt.figure(figsize=(10,6))

plt.scatter(risk_return_df['Risk'],
            risk_return_df['Returns'])

for i in risk_return_df.index:
    plt.text(risk_return_df['Risk'][i],
             risk_return_df['Returns'][i],
             i)
    plt.xlabel("Risk")
plt.ylabel("Average Return")
plt.title("Risk vs Return Analysis")

plt.show()
```

📊  TOP GAINERS & LOSERS

```python
total_return = ((df.iloc[-1] - df.iloc[0]) /
                df.iloc[0]) * 100

top_gainers = total_return.sort_values(ascending=False)

print("\nTop Gainers:")
print(top_gainers.head())

print("\nTop Losers:")
print(top_gainers.tail())

```
